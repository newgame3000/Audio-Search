```python
import matplotlib.pyplot as plt
import math
```

```python
a = []
ph = []
f = open('gr_f.txt', 'r')
k = 0

for line in f:
    
    
    if k == 0:
        a.append(float(line))
        k = 1
    else:
        ph.append(float(line))
        k = 0

gr_f = []
for i in range(len(a)):
    gr_f.append(a[i] * math.cos(ph[i]))
f.close()
```

```python
x = []
t = 0

for i in range(len(a)):
    x.append(t)
    t += (44100 / 4096)
```

## График сигнала

```python
f2 = open('gr.txt', 'r')
gr = []
k = 0
for line in f2:
    k += 1
    if k > 40000:
        continue
    gr.append(float(line))
    
len(gr)
f2.close()
```

```python
plt.plot(range(len(gr)), gr)
len(gr)
```
